```python
import pandas as pd
```

```python
flowers = pd.read_csv('iris_data.txt',",",header=None)
```

```python
flowers.describe()
```

```python
flowers
```

```python
flowers.columns=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Species']
```

```python
flowers
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flowers.plot(kind='scatter', x='Sepal width', y='Sepal length');
plt.title('Sepal length vs. Sepal width')
plt.show()
```

```python
flowers.plot(kind='scatter', x='Petal width', y='Sepal length');
plt.title('Sepal length vs. Petal width')
plt.show()
```

```python
flowers.plot(kind='scatter', x='Petal length', y='Sepal length');
plt.title('Sepal length vs. Petal length')
plt.show()
```

```python
flowers.plot(kind='scatter', x='Petal width', y='Sepal width');
plt.title('Sepal width vs. Petal width')
plt.show()
```

```python
flowers.plot(kind='scatter', x='Petal length', y='Sepal width');
plt.title('Sepal width vs. Petal length')
plt.show()
```

```python
flowers.plot(kind='scatter', x='Petal width', y='Petal length');
plt.title('Petal length vs. Petal width')
plt.show()
```

```python
flowers.boxplot(by='Species')
plt.show()
```
