```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import cluster, neighbors, datasets
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
flowers = pd.read_csv('iris_data.txt',",",header=None)
```

```python
flowers.describe()
```

```python
flowers
```

```python
flowers.columns=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Species']
```

```python
flowers
```

```python
speciesdata=flowers.groupby('Species')
SlPw=speciesdata.get_group('Iris-setosa').plot(kind='scatter', x='Petal width', y='Sepal length',c='Red', label='Iris-setosa', figsize=(12,8));
speciesdata.get_group('Iris-versicolor').plot(kind='scatter', x='Petal width', y='Sepal length',c='Green',label='Iris-versicolor',ax=SlPw);
speciesdata.get_group('Iris-virginica').plot(kind='scatter', x='Petal width', y='Sepal length',c='Blue', label='Iris-virginica',ax=SlPw);
SlPw.yaxis.tick_left()
SlPw.xaxis.tick_bottom()
plt.title('Sepal length vs. Petal width')
plt.show()
```
